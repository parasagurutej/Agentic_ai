```python
def grade(score):
  if(score>90):
    return'A'
  elif(score>=80):
    return 'B'
  else:
      return 'C'

#apply function
df['grade_letter']=df['score'].apply(grade)

print(df)
```

```python
result={
    True:"pass",
    False:"fail"
}
pas={'A':90,'B':80,'D':60}

df['pass']=df['grade'].map(pas)
df['result']=df['grade'].map(result)
print(df)
```

```python
df.drop('pass',axis=1,inplace=True)
```

```python
employees = pd.DataFrame({
    'emp_id': [1, 2, 3, 4],
    'name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'dept_id': [10, 20, 10, 30]
})

departments = pd.DataFrame({
    'dept_id': [10, 20],
    'dept_name': ['Engineering', 'HR']
})
```

```python
#inner join
pd.merge(employees,departments,on='dept_id',how='inner')
```

```python
#inner join
pd.merge(employees,departments,on='dept_id',how='outer')
```

```python

pd.merge(employees,departments,on='dept_id',how='left')
```

```python

pd.merge(employees,departments,on='dept_id',how='right')
```
